```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

# Supply Voltage Fitting

### Channel 1

```python
data1 = pd.read_csv("supply_callib_ch01_600.csv")
data1.rename(columns = {'0':'Percent', '1':'Input',
                       '2':'Output', '3':'SD'}, inplace = True)
data1 = data1.drop('Unnamed: 0', axis=1)
print(data1.shape)
data1.head()
```

```python
#load data in numpy arrays

percent1  = np.array(data1['Percent'])
input_v1  = np.array(data1['Input'])
output_v1 = np.array(data1['Output'])
std1      = np.array(data1['SD'])
```

```python
fig, ax = plt.subplots(2, 2, figsize=(2.5*6.4, 2.5*4.8))

ax[0][0].scatter(percent1[:150], input_v1[:150], color='red', label='input', s=0.05)
ax[0][0].scatter(percent1[:150], output_v1[:150], color='blue', label='output', s=0.05)
ax[0][0].set_xlabel("Percentage (0 to 1)")
ax[0][0].set_ylabel("Voltage (in V)")
ax[0][0].legend()

ax[0][1].scatter(percent1[150:300], input_v1[150:300], color='red', label='input', s=0.05)
ax[0][1].scatter(percent1[150:300], output_v1[150:300], color='blue', label='output', s=0.05)
ax[0][1].set_xlabel("Percentage (0 to 1)")
ax[0][1].set_ylabel("Voltage (in V)")
ax[0][1].legend()

ax[1][0].scatter(percent1[300:450], input_v1[300:450], color='red', label='input', s=0.05)
ax[1][0].scatter(percent1[300:450], output_v1[300:450], color='blue', label='output', s=0.05)
ax[1][0].set_xlabel("Percentage (0 to 1)")
ax[1][0].set_ylabel("Voltage (in V)")
ax[1][0].legend()

ax[1][1].scatter(percent1[450:], input_v1[450:], color='red', label='input', s=0.05)
ax[1][1].scatter(percent1[450:], output_v1[450:], color='blue', label='output', s=0.05)
ax[1][1].set_xlabel("Percentage (0 to 1)")
ax[1][1].set_ylabel("Voltage (in V)")
ax[1][1].legend()

fig.suptitle('Input and Output voltages for different percentages (Channel 1)', fontsize=16);
```

```python
difference1 = output_v1 - input_v1
plt.figure(figsize=(2*6.4, 2*4.8))
plt.scatter(percent1, difference1, color='blue', label='difference = output - input', s=0.1)
plt.xlabel("Percentage (0 to 1)")
plt.ylabel("Voltage (in V)")
plt.title("Difference in Output and Input voltage (Channel 1)")
plt.legend();
```

### We see in the above plots that the output voltage is greater than the input voltage and then linearly decreases, thereby becoming lesser than the input voltage. The output voltage does however follow a linear trend so a linear fit should suffice.

```python
plt.figure(figsize=(2*6.4, 2*4.8))
plt.scatter(input_v1, std1, color='blue', label='standard deviation', s=0.5)
plt.xlabel("Input Voltage (in V)")
plt.ylabel("Voltage (in V)")
plt.title("Standard deviation wrt different supply voltages (Channel 1)", fontsize=16)
plt.legend();
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

x_arr = percent1.reshape(-1,1)
degree = 1
lin_reg_model1 = LinearRegression()
polyreg1 = make_pipeline(PolynomialFeatures(degree), lin_reg_model1)
a = polyreg1.fit(x_arr, output_v1)

plt.figure(figsize=(2*6.4, 2*4.8))
plt.plot(x_arr, polyreg1.predict(x_arr), color='red', label='fit')
plt.scatter(percent1, output_v1, color='blue', label='output voltage', s=0.5)
plt.xlabel("Percent (from 0 to 1)")
plt.ylabel("Voltage (in V)")
plt.title("Fit plotted with Output Voltage data (Channel 1)", fontsize=16)
plt.legend();
```

```python
from sklearn.linear_model import LinearRegression

model1 = LinearRegression(fit_intercept=True)
model1.fit(X = data1[["Percent"]], y = data1["Output"])
print("Intercept: ", model1.intercept_)
print("Slope: ", model1.coef_)
```

```python
plt.scatter(percent1, output_v1 - model1.predict(x_arr))
plt.axhline(0, c='black', linewidth=1)
plt.xlabel("Percent (from 0 to 1)")
plt.ylabel("Residuals $Y - \hat{Y}$")
plt.title("Fit plotted with Output Voltage data (Channel 1)", fontsize=16);
```

# Thermal Calibration (for constant supply voltage of $-1V$)

### Channel 1

```python
thermal1 = pd.read_csv("temperature_ch01.csv")
thermal1.rename(columns = {'Unnamed: 0':'hour', '0':'minutes:seconds',
                       '1':'measured voltage'}, inplace = True)
print(thermal1.shape)
thermal1.head()
```

```python
hours = np.array(thermal1["hour"].unique())
mean_temp = np.array(thermal1.groupby(["hour"])["measured voltage"].median())
std_temp  = np.array(thermal1.groupby(["hour"])["measured voltage"].std())

plt.scatter(hours, mean_temp, s = 20, c = 'red')
plt.errorbar(hours, mean_temp, yerr = std_temp, fmt = '.', c = 'black')
plt.xlabel("Hours (military time)")
plt.ylabel("Mean voltage per hour")
plt.title("Mean voltage per hour");
print(f"Mean standard deviation: {np.mean(std_temp)}")
```

```python
plt.hist(thermal1["measured voltage"], bins=50, label="Distribution", color='pink')

plt.xticks(rotation=90)
plt.xlabel("Measured Voltage (in V)")
plt.ylabel("Counts")
mu, sigma = np.mean(thermal1["measured voltage"]), np.std(thermal1["measured voltage"])
print(f"Mean: {mu}, Standard deviation: {sigma}")
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 0.17*stats.norm.pdf(x, mu, sigma), linestyle='dashed', linewidth=1.5, color='red', label="Gaussian fit")
plt.axvline(x = mu, color = 'b', label = 'Mean')
plt.legend()
plt.title("Voltage distribution");
```
