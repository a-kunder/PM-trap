```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))

x  = np.linspace(0, 100, 100)
y1 = gaussian(x[:50], 25, 5)
y2 = gaussian(x[50:], 75, 5)
y  = np.append(y1, y2)
plt.plot(x, y);
```
